```python
import hashlib

# 1. The Original Record
data = "Payment Approved: $1,000"
print("Original Hash:", hashlib.sha256(data.encode()).hexdigest())

# 2. The 'Tampered' Record (we add a single period at the end)
tampered = "Payment Approved: $1,000."
print("Tampered Hash:", hashlib.sha256(tampered.encode()).hexdigest())
```

```python
  import hashlib

print("--- STEP 1: THE BUSINESS SEAL ---")
# 'Issue' a payment amount
original_data = input("Enter the original payment amount (eg $500): ")
original_hash = hashlib.sha256(original_data.encode()).hexdigest()

print(f"Blockchain Digital Seal: {original_hash}")
print("\n" + "-"*30 + "\n")

print("--- STEP 2: THE AUDIT CHECK ---")
# Ask another classmate to 'verify' the payment
check_data = input("Enter the amount to verify: ")
check_hash = hashlib.sha256(check_data.encode()).hexdigest()

print(f"Current Data Hash: {check_hash}")

# The moment of truth
if original_hash == check_hash:
    print("\n✅ VERIFIED: This data is authentic and hasn't been changed.")
else:
    print("\n❌ ALERT: Data Tampering Detected! The fingerprints do not match.")
```
